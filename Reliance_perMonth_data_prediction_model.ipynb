```python
import pandas as pd
import seaborn as sns
%matplotlib inline
#df=C:\Users\BESTSELLER\Desktop\coding\RELIANCE_Stock_data.xls
```

```python
df= pd.read_excel(r'C:\Users\BESTSELLER\Desktop\coding\RELIANCE_Stock_data.xls')
```

```python
df
```

```python
feature_cols=['Open','High','Low']
x=df[feature_cols]
y=df['Close']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(x, y, test_size=0.5)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
linreg= LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred=linreg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import metrics
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
data= X_test[['Open','High','Low']]
data['actual_closing']=y_test
data['predicted_closing_price']=y_pred
data
```

```python
import plotly_express as px
figg=px.scatter(data, x='Open', y='predicted_closing_price', hover_data=['actual_closing'])
figg.show()
```
